```python
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler
```

```python
## create 2 empty list for storing the smaple & lable data for traing 

train_sample = []
train_lable = []
```

```python
for i in range(1000):
    
    younger_age = randint(13,64)
    train_sample.append(younger_age)
    train_lable.append(0)
    
    older_age = randint(65,100)
    train_sample.append(older_age)
    train_lable.append(1)
```

```python
train_lable
```

```python
train_sample
```

```python
train_lable = np.array(train_lable)

train_sample = np.array(train_sample)
```

```python
train_lable
```

```python
train_sample
```

```python
scaler = MinMaxScaler()
scaler_train_sample = scaler.fit_transform(train_sample.reshape(-1,1))
```

```python
scaler_train_sample
```

```python
import keras
from keras.layers import Dense
from keras.models import Sequential
from keras.optimizers import Adam
```

```python
model = Sequential([Dense(units = 16, input_dim = 1, activation= 'relu', ),
                    Dense(32, activation= 'relu'),
                    Dense(2, activation= 'softmax')])
```

```python
model.summary()
```

```python
## compile the model

model.compile(Adam(learning_rate= 0.001), loss = 'sparse_categorical_crossentropy', metrics= ['accuracy'])
```

```python
model.fit(train_sample, train_lable, batch_size= 10, epochs= 10)
```

```python
## for test dataset 

test_samples = []
test_labels = []

for i in range(500):
    
    younger_age = randint(13,64)
    test_samples.append(younger_age)
    test_labels.append(0)
    
    older_age = randint(65,100)
    test_samples.append(older_age)
    test_labels.append(1)
```

```python
test_labels = np.array(test_labels)

test_samples = np.array(test_samples)
```

```python
scaler_test_samples = scaler.fit_transform(test_samples.reshape(-1,1))
```

```python
scaler_test_samples
```

```python
## predict the value using test samples

test_sample_output = model.predict(test_samples, batch_size= 10)
```

```python
model.evaluate(test_samples,test_labels)[1]
```
